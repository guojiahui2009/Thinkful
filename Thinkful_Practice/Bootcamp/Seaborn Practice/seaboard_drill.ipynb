```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df1 = pd.DataFrame(pd.read_csv("world.csv"))
df1.head(5)
```

```python
df2 = pd.DataFrame(pd.read_csv("nba_salary.csv"))
df2.rename(columns={'season17_18': 'salary', 'Tm':'team'}, inplace = True)
df2.head(5)
```

# Choose one variable and plot that variable four different ways.

salary of nba players

```python
sb.distplot(df2.salary,bins=20)
```

```python
sb.boxplot(df2.salary)
```

```python
sb.kdeplot(df2.salary,shade=1)
```

```python
sb.rugplot(df2.salary)
```

# Choose two continuous variables, and plot them three different ways.

df1 birthrate deathrate

```python
sb.jointplot(x = "Population", y = "Area(sq.mi.)", data = df1, kind = "kde")
```

```python
sb.lmplot(x = "Population", y = "Area(sq.mi.)", data = df1, fit_reg = 1, scatter_kws={'alpha':0.4})
```

```python
g = sb.jointplot(x = "Population", y = "Area(sq.mi.)", data = df1, kind="kde", color="g")
g.plot_joint(plt.scatter, c="b", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$Population$", "$Area$");
```

# Choose one continuous variable and one categorical variable, and plot them six different ways.

continuous salary, categorical team

```python
sb.stripplot(x="salary", y="team", data=df2)
```

```python
sb.swarmplot(x="salary", y="team", data=df2)
```

```python
sb.boxplot(x="salary", y="team", data=df2)
```

```python
sb.violinplot(x="salary", y="team", data=df2)
```

```python
sb.barplot(x="salary", y="team", data=df2)
```

```python
sb.violinplot(x="salary", y="team", data=df2, innner="stick")
```
