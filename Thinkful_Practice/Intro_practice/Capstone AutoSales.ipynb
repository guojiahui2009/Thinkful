```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(pd.read_csv("TOTALSA.csv"))
```

```python
#extract year and month values from date
dates = df["DATE"]
year = []
month = []
for date in dates:
    mylist = date.split("-")
    year.append(int(mylist[0]))
    month.append(int(mylist[1]))
df["Year"] = year
df["Month"] = month
df
```

```python
plt.plot(df["DATE"],df["TOTALSA"])
plt.ylabel("Sales in millions")
plt.title("Auto Sales by month")
plt.show()
```

```python
# average monthly sales figures
df["TOTALSA"].mean()
```

```python
#is there a particular month in the year people like to buy cars?
df.groupby("Year").aggregate(np.sum)["TOTALSA"]
#df.groupby("Year").aggregate(np.mean)
```
