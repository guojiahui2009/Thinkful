# Description
this data set is from the Federeal Reserve's Economic Data of the Federal reserve bank of Saint Louis. The raw data contains the monthly autosales figures in millions of units sold from January 1976 to April 2018. 

https://fred.stlouisfed.org/series/TOTALSA


```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

```python
# read data from file
df = pd.DataFrame(pd.read_csv("TOTALSA.csv"))
df.head(5)
```

```python
#extract year and month values from date
dates = df["DATE"]
year = []
month = []
for date in dates:
    mylist = date.split("-")
    year.append(int(mylist[0]))
    month.append(int(mylist[1]))
df["Year"] = year
df["Month"] = month

# summary of total sales column
df["TOTALSA"].describe()
```

```python
#np.histogram(df["TOTALSA"], bins = 15)
plt.hist(df["TOTALSA"], bins = np.histogram(df["TOTALSA"])[1]) 
plt.title("Monthly Sales Histogram") 
plt.ylabel("count")
plt.xlabel("Sales in Millions of units")
plt.show()
```

From the histogram the sales data does not look normal, and appears skewed to the left.
The fact that these data are generated over time may be the reason for abnormality. 

```python
# plot total sales through time
plt.plot(df["TOTALSA"])
plt.ylabel("Sales in millions")
plt.title("Auto Sales by month 1976 - 2018")
plt.show()
```

* please note the x axis is plotting through time, I had trouble displaying the dates in a presentable manner

there are some noticeable fluctuations along with what apear to be cyclical ups and downs in the data

```python
#average monthly sales throughout the years
#df.groupby("Year").aggregate(np.sum)["TOTALSA"]
df_average = pd.DataFrame()
df_average["sales"] = df.groupby("Year").aggregate(np.mean)["TOTALSA"]
plt.plot(df_average)
plt.title("Average Monthly Sales/Year")
plt.ylabel("Millions in sales")
plt.xlabel("Year")
plt.show()
```

when displayed in average monthly sales by year, the plot became smoother, but kept the general shape of the major fluctuations 

```python
#total sales throughout the years
#omit year 2018 in sum calculations
df_temp = df[df["Year"] != 2018]
df_sum = pd.DataFrame()
df_sum["TOTALSA"] = df_temp.groupby("Year").aggregate(np.sum)["TOTALSA"]
plt.plot(df_sum["TOTALSA"])
plt.title("Total Anual Sales")
plt.ylabel("Millions in sales")
plt.xlabel("Year")
plt.show()
```

total annual sales plot in sync with average monthly plot... as it's just 12 times in volumes, only difference is in the y axis

# Are there seasonal behaviors that are consistent annually?

```python
# average sales by month
plt.plot(df.groupby("Month").aggregate(np.mean)["TOTALSA"])
plt.title("Average Sales by Month")
plt.show()
```

looking at the average sales by month, it seems that the months of august, september, and december are most successful in terms of auto sales in units

```python
# here we are going to isolate the month in each year that had the highest sales
df = df.set_index(['Year','Month'])
indices = df.groupby('Year').idxmax()['TOTALSA']
df2 = df.loc[indices,:]
index_sets = df2.index.values
months = np.zeros(shape = (12,1))
for set in index_sets:
    months[set[1]-1] += 1

#load the frequency of highest sales of each month into a new dataframe with corresponding indexes
df_month_max_frequency = pd.DataFrame(index = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                                               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"] )
df_month_max_frequency["Frequency at Max"] = months
df_month_max_frequency.sort_values(["Frequency at Max"], ascending = False).plot(kind = "bar")
plt.show()
```

December seems to be people's favorite time of the year to purchase cars, 
which could be related to car sellers capitalizing on the holiday season and providing large incentives to prompt buyers to act. 

August and September are the two months where a push is seen, which could be partly explained by fiscal year of auto dealers actually normally begin in June, and the dramatic rise as a product of hard pushes in the first two to three months to meet new sales targets. 
Additionally, graduates from high school and colleges enter the labor force and could be in need of vehicles as their principle means of transportation in many areas, helping to drive up the sales. 

# Got Oil?

It seems that the massive dips and rises throughout the past four decades of autosales has some relationship with the price of oil. 

I have gathered some data on monthly oil prices from http://www.forecast-chart.com/chart-crude-oil.html
and will be testing to see if there is indeed a correlation

```python
#import data and align with sales data
df_oil = pd.DataFrame(pd.read_csv("oil_prices.csv"))

dates = df_oil["date"]
year = []
month = []
for date in dates:
    mylist = date.split("-")
    year.append(int(mylist[0]))
    month.append(int(mylist[1]))
df_oil["Year"] = year
df_oil["Month"] = month
df_oil["Sales"] = df[df["Year"] < 2018]["TOTALSA"]
df_oil.head(5)
```

```python
# use scatter plot for sales over prices to see if any relationshio can be found
plt.scatter(df_oil["price"], df_oil["Sales"])
plt.title("Sales over Oil Price")
plt.ylabel("Sales in Millions units/month")
plt.xlabel("Dollars per Barrel")
plt.show()
```

there is minimal resemblence of a negative linear relationship

```python
coef = np.corrcoef(df_oil["price"], df_oil["Sales"])[0][1]
coef
```

the correlation coefficient reaffirms the findings in the scatter plot in which only a very very small negative relationshio can be deducted. 

This is not surprising as autosales can be an complex topic where it is associated with many other variables other than oil prices.
