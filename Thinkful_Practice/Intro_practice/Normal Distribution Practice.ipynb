```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1 Poisson 

```python
poisson = np.random.poisson(10, 100)
plt.hist(poisson)

#add lines
def addlines(poisson):
    plt.axvline(poisson.mean(), color = 'y', linestyle = 'solid', linewidth = 2)
    plt.axvline(poisson.mean() + poisson.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
    plt.axvline(poisson.mean() - poisson.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
    pass

addlines(poisson)

plt.show()
```

looks like its skewed left

# 2 Beta

```python
beta = np.random.beta(1,2,100)
plt.hist(beta)
addlines(beta)
plt.show()
```

# 3 Gamma

```python
gamma = np.random.gamma(4, 1, 100)
plt.hist(gamma)
addlines(gamma)
plt.show()
```

# 4 Uniform

```python
uniform = np.random.uniform(0,20,100)
plt.hist(uniform)
addlines(uniform)
plt.show()
```

# 5 Power 

```python
power = np.random.power(10,100)
plt.hist(power)
addlines(power)
plt.show()
```

# 6 Pareto

```python
pareto = np.random.pareto(5,100)
plt.hist(pareto)
addlines(pareto)
plt.show()
```
